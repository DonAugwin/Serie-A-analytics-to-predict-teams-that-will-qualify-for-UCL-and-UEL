```python
train_df = df[(df['SEASON'] < 2020) | (df['SEASON'] > 2021)]
test_df = df[(df['SEASON'] >= 2020) & (df['SEASON'] <= 2021)]
X_train = train_df.drop(['Big League'], axis=1)
y_train = train_df['Big League']
X_test = test_df.drop(['Big League'], axis=1)
y_test= test_df['Big League']
```

```python
X_train=X_train.drop(['Squad','SEASON'],axis=1)
test_team=X_test['Squad']
X_test=X_test.drop(['Squad','SEASON'],axis=1)
```

```python
#Finding the suitable model
```

```python
from sklearn.linear_model import LogisticRegression, RidgeClassifier, Lasso
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score
from sklearn.ensemble import StackingClassifier
```

```python
estimators = [
    ('nb', GaussianNB()),
    ('svc', SVC(kernel='linear', probability=True)),
    ('knn', KNeighborsClassifier()),
    ('logistic', LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1300)),
    ('ridge', RidgeClassifier()),
]
```

```python
meta_clf = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1300)
```

```python
stacked_clf = StackingClassifier(
    estimators=estimators, final_estimator=meta_clf
)
```

```python
stacked_clf.fit(X_train, y_train)
```

```python
y_pred = stacked_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
report = classification_report(y_test, y_pred)

print('Accuracy:', accuracy)
print('F1 score:', f1)
print('Classification report:\n', report)
```

```python

# Obtain predicted probabilities for each class
y_score_stacked = stacked_clf.predict_proba(X_test)

# Binarize the target variable for each class
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

# Compute ROC curve and AUC for each class
n_classes = y_test_bin.shape[1]
fpr_stacked = {}
tpr_stacked = {}
roc_auc_stacked = {}
for i in range(n_classes):
    fpr_stacked[i], tpr_stacked[i], _ = roc_curve(y_test_bin[:, i], y_score_stacked[:, i])
    roc_auc_stacked[i] = auc(fpr_stacked[i], tpr_stacked[i])

# Compute micro-average ROC curve and AUC
fpr_stacked["micro"], tpr_stacked["micro"], _ = roc_curve(y_test_bin.ravel(), y_score_stacked.ravel())
roc_auc_stacked["micro"] = auc(fpr_stacked["micro"], tpr_stacked["micro"])

# Plot ROC curves for each class and micro-average
plt.figure()
lw = 2
colors = cycle(['red', 'blue', 'green'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr_stacked[i], tpr_stacked[i], color=color, lw=lw, label='Class %d (AUC = %0.2f)' % (i, roc_auc_stacked[i]))

plt.plot(fpr_stacked["micro"], tpr_stacked["micro"], color='deeppink', lw=lw, linestyle='--', label='Micro-average (AUC = %0.2f)' % roc_auc_stacked["micro"])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Obtain predicted probabilities for each class
y_score_stacked = stacked_clf.predict_proba(X_test)

# Binarize the target variable for each class
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

# Compute ROC curve and AUC for each class
n_classes = y_test_bin.shape[1]
fpr_stacked = {}
tpr_stacked = {}
roc_auc_stacked = {}
for i in range(n_classes):
    fpr_stacked[i], tpr_stacked[i], _ = roc_curve(y_test_bin[:, i], y_score_stacked[:, i])
    roc_auc_stacked[i] = auc(fpr_stacked[i], tpr_stacked[i])

# Compute macro-average ROC curve and AUC
fpr_stacked["macro"] = np.unique(np.concatenate([fpr_stacked[i] for i in range(n_classes)]))
tpr_stacked["macro"] = np.zeros_like(fpr_stacked["macro"])
for i in range(n_classes):
    tpr_stacked["macro"] += np.interp(fpr_stacked["macro"], fpr_stacked[i], tpr_stacked[i])
tpr_stacked["macro"] /= n_classes
roc_auc_stacked["macro"] = auc(fpr_stacked["macro"], tpr_stacked["macro"])

# Plot ROC curves for each class and macro-average
plt.figure()
lw = 2
colors = cycle(['red', 'blue', 'green'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr_stacked[i], tpr_stacked[i], color=color, lw=lw, label='Class %d (AUC = %0.2f)' % (i, roc_auc_stacked[i]))

plt.plot(fpr_stacked["macro"], tpr_stacked["macro"], color='deeppink', lw=lw, linestyle='--', label='Macro-average (AUC = %0.2f)' % roc_auc_stacked["macro"])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Get predicted probabilities for each class
y_score_svm = svm.predict_proba(X_test)

# Convert labels to binary format
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))

# Compute ROC curve and AUC for each class
n_classes = y_test_bin.shape[1]
fpr_svm = {}
tpr_svm = {}
roc_auc_svm = {}
for i in range(n_classes):
    fpr_svm[i], tpr_svm[i], _ = roc_curve(y_test_bin[:, i], y_score_svm[:, i])
    roc_auc_svm[i] = auc(fpr_svm[i], tpr_svm[i])

# Compute micro-average ROC curve and AUC
fpr_svm["micro"], tpr_svm["micro"], _ = roc_curve(y_test_bin.ravel(), y_score_svm.ravel())
roc_auc_svm["micro"] = auc(fpr_svm["micro"], tpr_svm["micro"])

# Plot ROC curves for each class and micro-average
plt.figure()
lw = 2
colors = cycle(['red', 'blue', 'green'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr_svm[i], tpr_svm[i], color=color, lw=lw, label='Class %d (AUC = %0.2f)' % (i, roc_auc_svm[i]))
    
plt.plot(fpr_svm["micro"], tpr_svm["micro"], color='deeppink', lw=lw, linestyle='--', label='Micro-average (AUC = %0.2f)' % roc_auc_svm["micro"])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
